```python
from collections import deque

class Node:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data

def breadth_first_search(root):
    if root is None:
        return
    
    queue = deque()
    queue.append(root)
    
    while queue:
        node = queue.popleft()
        print(node.data, end=' ')
        
        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)

def create_binary_tree():
    n = int(input("Enter the number of nodes in the tree: "))
    if n <= 0:
        print("Invalid number of nodes.")
        return None
    
    root_data = int(input("Enter the value of the root: "))
    root = Node(root_data)
    queue = deque()
    queue.append(root)
    
    while len(queue) > 0:
        current_node = queue.popleft()
        
        left_data = input(f"Enter the value of the left child of {current_node.data} (or 'None'): ")
        if left_data.lower() != 'none':
            current_node.left = Node(int(left_data))
            queue.append(current_node.left)
        
        right_data = input(f"Enter the value of the right child of {current_node.data} (or 'None'): ")
        if right_data.lower() != 'none':
            current_node.right = Node(int(right_data))
            queue.append(current_node.right)
    
    return root

# Create the binary tree
root = create_binary_tree()

# Perform breadth-first search and print the nodes
print("BFS of the tree is:")
breadth_first_search(root)
```
